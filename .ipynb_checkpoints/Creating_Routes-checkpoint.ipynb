```python
# create set with all edges, not fitting ones are removed later
E_temp = set([(i,j) for i in N for j in N if j not in [i]])
distance = {(i,j): calc_distance(coord[i], coord[j]) for (i,j) in E_temp}

# for each node check all edges to other nodes, from furthest to closest, if they are too similar to other edges
for i in N:
    dist_to_j = {j: distance[i,j] for j in N if j not in [i]}
       
    while len(dist_to_j) > 1:
        furthest = max(dist_to_j, key=dist_to_j.get) # node that is the furthest away from current node
        del dist_to_j[furthest] # remove from list of edges to check
        # vector to the furthest remaining node node
        vector_furthest = np.array([coord[i][0] - coord[furthest][0], coord[i][1] - coord[furthest][1]]) 
        # angles to all the other nodes
        angles = {j: calc_angle(vector_furthest, np.array([coord[i][0] - coord[j][0], coord[i][1] - coord[j][1]])) 
                  for j in dist_to_j}

        for j in angles:          
            if abs(angles[j]) < 20: # angle to small
                E_temp.remove((i, furthest))              
                break
            elif abs(angles[j]) < 60:     # angle in a medium size, check if the detour would be too big to keep it           
                short_dist_j_to_all, temp = dijkstra(distance, j)
                dist_i_j_furthest = distance[i,j] + short_dist_j_to_all[furthest]
                if  distance[i,furthest] * 1.5  > dist_i_j_furthest:
                    E_temp.remove((i, furthest))
                    break
                

# create set with final edges                
E = set()
for (i,j) in E_temp: # add all the reverse arcs
    E.add((j,i))
    E.add((i,j))
          
# find the closest node for customers and depots and add arcs to and from this node to E
for c in CD:
    dist_cus = {n: calc_distance(coord[c], coord[n]) for n in N}
    n = min(dist_cus, key=dist_cus.get)
    E.add((c, n))
    E.add((n, c))

# recalculate all the distances        
distance = {(i,j): calc_distance(coord[i], coord[j]) for (i,j) in E}
driving_time = calc_driving_time(distance, velocity, len_step)
                         
print('Data read from excel file')
```

```python
# ---create inital routes with a savings heuristik--- #


initial_routes = []

for d in D:
    # create a route for each customer
    routes = {i: Route(ID=i, nodes=[d, allocation[d][i], d], factor_fuel= cost_fuel, factor_labor=cost_labor ) for i in range(len(allocation[d]))} # routes at a depot, to be able to delete it later with the key
    for k, v in routes.items():
        v.update_nodes(V)
        v.update_route(V, shortest_drive_times, shortest_distance)
    nodes_at_depot = [d] + allocation[d].copy() # the depot and all customers allocated to it
    costs = {(i,j): shortest_distance[i,j] for i in nodes_at_depot for j in nodes_at_depot}
    savings = {(i,j): (round(costs[i, d] + costs[d, j] - costs[i, j],2)) for i in allocation[d]  for j in allocation[d] if i not in [j]}

    
    while len(savings) > 0: 
        # find biggest saving
        (n1,n2) = max(savings, key=savings.get)
        # route_update = True
        
        # update routes
        r1 = V[n1].route
        r2 = V[n2].route
        if routes[r1].demand + routes[r2].demand < capacity_truck:
            pos1 = routes[r1].nodes.index(n1)
            pos2 = routes[r2].nodes.index(n2)
            # XOR operator, if NOT exaclty one of the positions is 1, the second route has to be reversed
            if not ((pos1==1) ^ (pos2==1)):
                routes[r2].nodes.reverse()
            routes[r1].nodes =routes[r1].nodes[0:-1] + routes[r2].nodes[1:] # new combined route
            del routes[r2] # delete old route
            
            # updating route information
            routes[r1].update_route(V, shortest_drive_times, shortest_distance)
            routes[r1].update_nodes(V)
                      
            # update costs with the new routes        
            costs[d,n2] = costs[d,n1] + costs[n1,n2]
            costs[n1,d] = costs[n1,n2] + costs[n2,d]

            # recalculate savings
            savings = dict()
            for r1 in routes.values():
                #start and end node on current_tour
                current_nodes = (r1.nodes[1], r1.nodes[-2])
                for n1 in current_nodes:
                # find all start or end notes on the other routes from this depot          
                    other_nodes = set()
                    for r2 in routes.values():
                        if r1.ID != r2.ID:
                            other_nodes.add(r2.nodes[1]) # add first customer on route to nodes with possible savings 
                            other_nodes.add(r2.nodes[-2]) # add last customer on route to nodes with possible savings
                    # calculate possible savings
                    for n2 in other_nodes:
                        savings[n1, n2] = round(costs[n1, d] + costs[d, n2] - costs[n1, n2], 2)
        else: # combined demand would be over the capacity of a truck, so not possible --> delete saving
            del savings[n1,n2]

    # no more savings in for this depot possible, save routes to final routes 
    initial_routes = initial_routes + list(routes.values())

  
routes = {i: initial_routes[i] for i in range(len(final_routes))}    
for k, v in routes.items():
    v.ID = k
    v.update_nodes(V)
   


# find the exact pathes for the routes
for r in routes.values():
    r.find_exact_route(predecessor_shortest_path)
    r.update_route()
print(routes) 
```
